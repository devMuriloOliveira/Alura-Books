```python
import requests

def traduzir_texto(texto, origem="en", destino="pt"):
    # 1ï¸âƒ£ Tenta primeiro com LibreTranslate
    try:
        url = "https://libretranslate.de/translate"
        payload = {
            "q": texto,
            "source": origem,
            "target": destino,
            "format": "text"
        }

        resposta = requests.post(url, data=payload, timeout=10)

        # Verifica status
        if resposta.status_code == 200:
            data = resposta.json()
            if "translatedText" in data:
                return data["translatedText"]
            else:
                print("âš ï¸ Resposta inesperada do LibreTranslate:", data)
        else:
            print(f"âš ï¸ Erro HTTP {resposta.status_code} no LibreTranslate")

    except Exception as e:
        print("âŒ Erro ao usar LibreTranslate:", e)

    # 2ï¸âƒ£ Fallback para MyMemory se o primeiro falhar
    try:
        url2 = "https://api.mymemory.translated.net/get"
        params = {
            "q": texto,
            "langpair": f"{origem}|{destino}"
        }

        resposta2 = requests.get(url2, params=params, timeout=10)
        if resposta2.status_code == 200:
            data2 = resposta2.json()
            return data2["responseData"]["translatedText"]
        else:
            print(f"âš ï¸ Erro HTTP {resposta2.status_code} no MyMemory")

    except Exception as e:
        print("âŒ Erro ao usar MyMemory:", e)

    return "NÃ£o foi possÃ­vel traduzir o texto."


# ğŸŸ¢ Texto base
texto = "Hello, world! How are you today?"

# ğŸŒ Idiomas para traduzir
idiomas = {
    "pt": "PortuguÃªs",
    "en": "InglÃªs",
    "ja": "JaponÃªs",
    "es": "Espanhol",
    "fr": "FrancÃªs"
}

print("Texto original:", texto)
print("\nğŸ”¤ TraduÃ§Ãµes:\n")

# ğŸ” Traduz para cada idioma
for codigo, nome in idiomas.items():
    traducao = traduzir_texto(texto, origem="en", destino=codigo)
    print(f"{nome} ({codigo}): {traducao}")
```
