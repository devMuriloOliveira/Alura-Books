```python


import requests


def traduzir_texto(texto, origem="en", destino="pt"):
    # 1Ô∏è‚É£ Tenta primeiro com LibreTranslate
    try:
        url = "https://libretranslate.de/translate"
        payload = {
            "q": texto,
            "source": origem,
            "target": destino,
            "format": "text"
        }


        resposta = requests.post(url, data=payload, timeout=10)


        # Verifica status
        if resposta.status_code == 200:
            data = resposta.json()
            if "translatedText" in data:
                return data["translatedText"]
            else:
                print("‚ö†Ô∏è Resposta inesperada do LibreTranslate:", data)
        else:
            print(f"‚ö†Ô∏è Erro HTTP {resposta.status_code} no LibreTranslate")


    except Exception as e:
        print("‚ùå Erro ao usar LibreTranslate:", e)


    # 2Ô∏è‚É£ Fallback para MyMemory se o primeiro falhar
    try:
        url2 = "https://api.mymemory.translated.net/get"
        params = {
            "q": texto,
            "langpair": f"{origem}|{destino}"
        }


        resposta2 = requests.get(url2, params=params, timeout=10)
        if resposta2.status_code == 200:
            data2 = resposta2.json()
            return data2["responseData"]["translatedText"]
        else:
            print(f"‚ö†Ô∏è Erro HTTP {resposta2.status_code} no MyMemory")


    except Exception as e:
        print("‚ùå Erro ao usar MyMemory:", e)


    return "N√£o foi poss√≠vel traduzir o texto."


# üü¢ Teste
texto = "Hello, world! How are you today?"
traducao = traduzir_texto(texto)


print("Texto original:", texto)
print("Tradu√ß√£o:", traducao)


//===============================================================//

from PIL import Image
import requests
from io import BytesIO
from transformers import pipeline

# Carregar modelo pr√©-treinado
classificador = pipeline("image-classification", model="microsoft/resnet-50")

# ‚úÖ URL de imagem v√°lida (Golden Retriever)
url_imagem = "https://static.wikia.nocookie.net/tkoc/images/0/09/Goku.png/revision/latest?cb=20130424212620&path-prefix=pt-br"

# Fazer download com cabe√ßalho
headers = {"User-Agent": "Mozilla/5.0"}
resposta = requests.get(url_imagem, headers=headers)

if resposta.status_code == 200:
    imagem = Image.open(BytesIO(resposta.content))
    imagem.show()

    resultados = classificador(imagem)

    print("Resultados da classifica√ß√£o:\n")
    for r in resultados:
        print(f"{r['label']} - Confian√ßa: {r['score']:.2%}")

    melhor = resultados[0]
    print(f"\nüèÜ Classe mais prov√°vel: {melhor['label']} ({melhor['score']:.2%} de confian√ßa)")

else:
    print(f"‚ùå Erro ao baixar a imagem. C√≥digo HTTP: {resposta.status_code}")
```
